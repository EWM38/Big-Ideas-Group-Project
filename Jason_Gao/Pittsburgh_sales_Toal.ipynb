```python
import pandas as pd
import os
import matplotlib.pyplot as plt

# Get the current working directory and construct the file path
file_path = os.path.join(os.getcwd(), 'Sales-transaction-2024-geocoded_with_neighborhood.csv')

# Load the CSV file
df = pd.read_csv(file_path)

# Convert SALEDATE to datetime (if it's not already)
df['SALEDATE'] = pd.to_datetime(df['SALEDATE'])

# Filter the data for SALEDATE > 2023 and SCHOOLCODE == 47
df_filtered = df[(df['SALEDATE'].dt.year > 2023) & (df['SCHOOLCODE'] == 47)]

# Group by neighborhood, then aggregate data (e.g., count the number of transactions)
grouped = df_filtered.groupby('neighborhood').size().reset_index(name='Count')

# Display the result
print(grouped)

# Plot the results (optional)
plt.figure(figsize=(10, 6))
plt.bar(grouped['neighborhood'], grouped['Count'])
plt.xlabel('Neighborhood')
plt.ylabel('Number of Transactions')
plt.title('Sales Transactions by Neighborhood')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
